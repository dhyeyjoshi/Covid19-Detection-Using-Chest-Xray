```python
import pandas as pd
import os 
import shutil
```

```python
#Create data for the positive samples(Git-hub)
FILE_PATH ="chestxray/metadata.csv"
IMAGES_PATH = "chestxray/images"
```

```python
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
TARGET_DIR= "Dataset/Covid"

if not os.path.exists(TARGET_DIR) :
    os.mkdir(TARGET_DIR)
    print("Covid folder is created")
```

```python
cnt = 0

for (i,row) in df.iterrows():
    if row["finding"]=="COVID-19" and row ["view"]=="PA" :
        filename=row["filename"]
        image_path = os.path.join(IMAGES_PATH,filename)
        image_copy_path = os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
       # print("Moving image" , cnt)
        cnt += 1 
```

```python
#Sampling of Images from kaagle 

import random 
KAAGLE_FILE_NAME="chest_xray_kaagle/train/NORMAL"
TARGET_NORMAL_DIR="Dataset/Normal"
```

```python
image_names = os.listdir(KAAGLE_FILE_NAME)
```

```python
image_names
```

```python
random.shuffle(image_names)
```

```python
for i in range(142) :
    
    image_name=image_names[i]
    image_path = os.path.join(KAAGLE_FILE_NAME,image_name)
    target_path = os.path.join(TARGET_NORMAL_DIR,image_name)
    shutil.copy2(image_path,target_path)
    print("Copying image",i)
```
